**Logistic Regression**

```python
X=data[['price']] # for sklearn, we must have two dimensions for independent variable, this explains the use of two brackets
```

```python
Y=data['Sold']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf_lrs=LogisticRegression()
```

```python
model=clf_lrs.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Thus Y=-0.03571865x + 0.614775116
# H(Y)==1 , if (Y>0.5)
#H(Y)==0  , if(Y<0.5)
```

```python
import statsmodels.api as sn
X_cons=sn.add_constant(X)
X_cons.head()
```

```python
import statsmodels.discrete.discrete_model as sm
```

```python
logit=sm.Logit(Y,X_cons).fit()
```

```python
logit.summary()
```

**Logistic Regression with Multiple Independent Variables**

```python
X=data.loc[:,data.columns!='Sold']
```

```python
Y=data['Sold']
```

```python
clf_lr=LogisticRegression()
clf_lr.fit(X,Y)
```

```python
clf_lr.coef_
```

```python
clf_lr.intercept_
```

```python
X_cons=sn.add_constant(X)
```

```python
logit=sm.Logit(Y,X_cons).fit()
```

```python
logit.summary()
```

**Predicting and confusion matrix**

```python
clf_lr.predict_proba(X)
```

```python
y_pred=clf_lr.predict(X)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y,y_pred)
```
