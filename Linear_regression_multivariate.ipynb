# Variance Inflation Factor

```python
# VIF is required to calculate multicorrelation
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
  
```

```python
pd.Series([variance_inflation_factor(X.values, i) 
               for i in range(X.shape[1])], 
              index=X.columns)
```

# Generally,a VIF greater than 5 indicates strong multicolliniearity. An infinite VIF indicates that there is perfect multicollinearity.

```python
#From the VIFs , we can observe there is strong multicollineariy between the movie Genres.
#Multicollinearity does not necessarily effect the predictions.
#Multicollinearity undermines the statistical significance of an independent variable.
#Here it is important to point out that multicollinearity does not affect the model's predictive accuracy. 
#The model should still do a relatively decent job predicting the target variable when multicollinearity is present.
```

# Predicting on test dataset

```python
#Predictions
linear_model.predict(X_test.iloc[:1,:])
```

```python
Y_test=pd.DataFrame(Y_test)
```

```python
Y_test.iloc[:5,:]
```

```python
X_test.shape
```

```python
error=[]
for i in range(X_test.shape[0]):
    j=i+1
    pred=int(linear_model.predict(X_test.iloc[i:j,:]))
    true=int(np.array(Y_test.iloc[i:j,:]))
    diff=pred-true
    error.append(diff)
```

```python
import matplotlib.pyplot as plt
```

```python
index=np.linspace(1,X_test.shape[0],X_test.shape[0])
```

```python
error[:2]
```

```python
plt.figure(figsize=(10,8))
plt.plot(index,error,c='r')
plt.xlabel("index_of_test_sample")
plt.ylabel("Error")
plt.show()
```

```python
#Mean squared error:
import math
mean_squared_error=0
for i in range(X_test.shape[0]):
    j=i+1
    pred=int(linear_model.predict(X_test.iloc[i:j,:]))
    true=int(np.array(Y_test.iloc[i:j,:]))
    diff=pred-true
    err=(diff*diff)
    mean_squared_error+=err
mean_squared_error=math.sqrt(mean_squared_error/X_test.shape[0])
```

```python
print("The mean squared error is:{}".format(mean_squared_error))
```
