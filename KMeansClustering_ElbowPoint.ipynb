**K Means Clustering algorithm:kmeans.png**

```python
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
style.use("fivethirtyeight")
```

```python
?make_blobs()
```

```python
#make blobs() generates sample points, around c centers
X,Y=make_blobs(n_samples=100,centers=4,cluster_std=1,n_features=2)
```

```python
#Plotting the data points
plt.scatter(X[:,0],X[:,1],s=20,color='b')
```

```python
?plt.scatter
```

```python
?KMeans()
```

```python
#cost function
cost=[]
for i in range(1,11):
  KM=KMeans(n_clusters=i,max_iter=500)
  KM.fit(X)

  cost.append(KM.inertia_)
plt.plot(range(1,11),cost,color='g',linewidth='3')
plt.xlabel("Value of K")
plt.ylabel("Squared error(cost)")
plt.show()
```

```python
#Normalized cost function
cost=[]
for i in range(1,11):
  KM=KMeans(n_clusters=i,max_iter=500)
  KM.fit(X)
  m=len(X)
  normalized_cost=KM.inertia_/m;
  cost.append(normalized_cost)
plt.figure(figsize=(20, 10))
plt.margins(x=0, y=0)
plt.plot(range(1,11),cost,color='g',linewidth='3')
plt.xlabel("Value of K")
plt.ylabel("Normalized Squared error(cost)")
plt.xticks(range(1,11))

plt.show()
```

```python
#As we can see, the value of K=3 corresponds to the elbow point.
```
