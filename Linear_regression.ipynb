```python
#defining cost function
def cal_cost(theta,X,Y):
  m=len(Y)#length of training examples
  pred=X.dot(theta)#making predictions
  cost_val=(1/2*m)*np.sum(pow((pred-Y),2))#According to definition
  return cost_val
```

```python
def gradient_descent(X,Y,theta,learning_rate,iter):
  m=len(Y)#length of training examples
  #Initializing the coefficients:
  theta_all=np.zeros((iter,X.shape[1]))
  #Keeping track of cost function
  cost_all=np.zeros(iter)
  for e in range(iter):
    pred=np.dot(X,theta)
    theta=theta-(1/m)*learning_rate*(X.T.dot((pred-Y)))
    theta_history[e,:]=theta.T#Keeping track of the coefficients on each iteration
    cost_history[e]=cal_cost(theta,X,Y)#Keeping track of cost function on each iteration
  return theta,cost_all,theta_all
```

```python
X=2*np.random.rand(100,1)
Y=4+3*X
```

```python
lr=0.001
n_iter=1000
theta=np.random.randn(2,1)
X_new=np.c_[np.ones((len(X),1)),X]#building the feature matrix
theta,cost_history,theta_history=gradient_descent(X_new,Y,theta,lr,n_iter)
```
